```python


# Get the uploaded file name (assuming it's the first file uploaded)
file_path = list(uploaded.keys())[0]

# Load all sheets in the Excel file into a dictionary
dfs = pd.read_excel(file_path, sheet_name=['100', '300', '500','700'])
```

```python
# Define different marker styles
markers = ['o', 's', '^', 'D', 'x']  # Add more markers if needed
colors = ['blue', 'orange', 'green', 'red','black']

# Iterate through each sheet (temperature)
for temp, df in dfs.items():
    # Create a new figure for each temperature
    plt.figure(figsize=(8,5))

    # Extract the aging time (assuming it's the first column)
    aging_time = df.iloc[:, 0].values  # Get the first column for aging time

    # Plot data for each material (assuming each material is a column after the first)
    for i, col in enumerate(df.columns[1:]):
        conductivity = df[col].values  # Get the hardness values for each material
        marker = markers[i % len(markers)]  # Cycle through marker styles if more than 5 materials
        # Corrected line: pass a single color from the list based on the index 'i'
        plt.plot(aging_time, conductivity, label=col, color=colors[i % len(colors)], marker=marker, markersize=6, linestyle='-', linewidth=2)

    # Set the title and labels

    plt.xlabel('Aging Time (hours)',fontsize=14, fontweight='bold')
    plt.ylabel('Electrical Conductivity (% IACS)',fontsize=14, fontweight='bold')
    plt.legend()
    plt.title(f'at {temp}C', fontsize=16, fontweight='bold')
    plt.grid(True, linestyle='--', alpha=0.4)
    # Set y-axis limits to 60 to 110 for consistency
    plt.ylim(85, 105)

    # Set x-axis to display the actual aging time values, starting from 0
    plt.xlim(left=0)  # Ensure the x-axis starts from 0



    # Display the plot
    plt.tight_layout()
    plt.show()
```

# 50% rolled composite with different temperature of aging


```python
filename = list(uploaded.keys())[0]

df = pd.read_excel(filename, sheet_name="only .05% Al2O3")
df
```

```python
x = df['Temp']
y = df['avg']
err = df['error']

plt.figure(figsize=(8,6))

# Labeling
plt.xlabel('Aging temperature (C)', fontsize=13)
plt.ylabel('Electrical Conductivity (IACS%)', fontweight='bold', fontsize=13)
plt.title(f'aging of 50% rolled sample of .05% alumina composite')

# --- Line diagram with error bars ---
plt.plot(x, y, marker='o', linestyle='-', linewidth=2)        # line + markers
plt.errorbar(x, y, yerr=err, fmt='o', capsize=5, ecolor='black')

# Y-axis limit
plt.ylim(98, 103)

# Grid
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
