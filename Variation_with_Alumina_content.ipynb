```python
# Install necessary packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab import files
```

```python
# Upload the Excel file manually
uploaded = files.upload()

# Get the filename
filename = list(uploaded.keys())[0]

df = pd.read_excel(filename, sheet_name="Base var. No roll aging ")
df

```

```python
x=df['x']
x_1 = [str(x) for x in df['x']] #convert x values into string for separate plot
y=df['Conductivity Avg']
err=df['err_C']
#colors = ['red', 'blue', 'green', 'orange']

plt.figure(figsize=(8,6)) #customize the size

#labeling the Axis
plt.xlabel('Weight percentage(w%) of Al$_2$O$_3$-Reinforcement in Copper matrix', fontsize=13)
plt.ylabel('Electrical Conductivity (IACS%)', fontweight='bold', fontsize=13)
#plt.title("Bar Chart with Standard Deviation", fontsize=16) #bar title

#plot the bar-graph
plt.bar(x_1, y,
        yerr= err,color='orange', width=0.7,
        capsize=8,
        error_kw={'elinewidth': 1, 'ecolor': 'black'} #error bar customization
        )

plt.ylim(80, 105)        # y-axis range


# Grid and display settings
plt.grid(True, linestyle='--', alpha=0.5)



plt.show()
```

```python
df.columns = df.columns.str.strip() # Clean column names

x=df['x']
x_1 = [str(x) for x in df['x']] #convert x values into string for separate plot
y=df['HV Avg']
err=df['err_H']
#colors = ['red', 'blue', 'green', 'orange']

plt.figure(figsize=(8,6)) #customize the size

#labeling the Axis
plt.xlabel('Weight percentage(w%) of Al$_2$O$_3$-Reinforcement in Copper matrix', fontsize=13)
plt.ylabel('Microhardness (HV)', fontweight='bold', fontsize=13)
#plt.title("Bar Chart with Standard Deviation", fontsize=16) #bar title

#plot the bar-graph
plt.bar(x_1, y,
        yerr= err,color='skyblue', width=0.7,
        capsize=8,
        error_kw={'elinewidth': 1, 'ecolor': 'black'} #error bar customization
        )

plt.ylim(50, 90)        # y-axis range


# Grid and display settings
plt.grid(True, linestyle='--', alpha=0.5)
```
